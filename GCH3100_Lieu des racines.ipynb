```python
# @title Click `Show code` in the code cell. { display-mode: "form" }

# INSTALLATION
# %pip install -q kaleido
# %pip install -q ipywidgets
# %pip install -q plotly
# %pip install -q scipy
# %pip install -q control
# %pip install -q control_plotly

# IMPORTATION DES LIBRAIRIES
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio
import control
from   ipywidgets import interactive
from   scipy import signal
from control_plotly import rlocus

# CONFIGURATION
pio.templates.default = "plotly_white"
s  = control.tf("s")

# RÉSOLUTION SYMBOLIQUE

# RÉSOLUTION NUMÉRIQUE

# PRÉPARATION POUR VISUALISATION

# VISUALISATION INTERACTIVE
def graphique_interactif(p1,z1):

       # SYSTÈME
       Y_Ysp = (s-z1) / ((s+1)*(s+3)*(s-p1)*(s+complex(0,1))*(s-complex(0,1)))

       # VISUALISATION 
       fig = rlocus(Y_Ysp)

       # VISUALISATION
       fig.update_layout(dict(     title=dict(text="pas très bon",
                                      font_size=16),

                                   xaxis=dict(title="$\\text{Re }(s^{-1})$",
                                          titlefont_size=16,
                                          linecolor="black",
                                          showline=True,
                                          mirror=True,
                                          linewidth=1,
                                          gridcolor='darkgray',
                                          gridwidth=1,
                                          zeroline=False,
                                          domain=[0,.8],
                                          range=[-10,10],
                                          ),
                     
                                   yaxis=dict(title="$\\text{Im }(s^{-1})$",
                                          titlefont_size=16,
                                          linecolor="black",
                                          showline=True,
                                          mirror=True,
                                          linewidth=1,
                                          zeroline=True,
                                          range=[-2,2],
                                          ),

                                   showlegend = True,
                                   legend=dict(orientation="v",
                                          yanchor="top",
                                          y=1,
                                          xanchor="center",
                                          x=1),

                                   width=500,
                                   height=450,                            
                                   ))
       
       

       return fig.show(renderer="svg")

interactive_plot = interactive(graphique_interactif, p1=(-10, 10, 0.1),z1=(-10, 10, 0.1))
output = interactive_plot.children[-1]
output.layout.height = '400px'
interactive_plot
```

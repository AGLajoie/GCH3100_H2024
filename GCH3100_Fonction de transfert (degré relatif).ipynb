```python
# @title Click `Show code` in the code cell. { display-mode: "form" }

# INSTALLATION
# %pip install -q kaleido
# %pip install -q ipywidgets
# %pip install -q pint
# %pip install -q plotly
# %pip install -q control

# IMPORTATION DES LIBRAIRIES
import numpy as np
import control as ct
import plotly.graph_objects as go
import plotly.io as pio
from scipy import signal
import sympy as sp
s = sp.symbols("s")

# CONFIGURATION
pio.templates.default = "plotly_white"

# SYSTÈME
def Visualisation(num, den, dom_x):
    x_vecteur = np.linspace(0,dom_x,1000)

    # Fonction de transfert
    G = signal.lti(num, 
                den)

    # Simulation
    _, y_vecteur = signal.step(G,T=x_vecteur)

    # VISUALISATION
    fig = go.Figure(go.Scatter(x=x_vecteur, y=y_vecteur))

    fig.update_layout(dict(title=dict(text="Fonction de transfert",
                                    font_size=16),

                        xaxis=dict(title="Temps [s]",
                                    titlefont_size=16,
                                    linecolor="black",
                                    showline=True,
                                    mirror=True,
                                    linewidth=1,
                                    gridcolor='darkgray',
                                    gridwidth=1,
                                    zeroline=False,
                                    #dtick = 2
                                    ),

                        yaxis=dict(title="Réponse [-]",
                                    titlefont_size=16,
                                    linecolor="black",
                                    showline=True,
                                    mirror=True,
                                    linewidth=1,
                                    gridcolor='darkgray',
                                    gridwidth=1,
                                    zeroline=False,
                                    #dtick = 2
                                    ),

                        coloraxis=dict(colorbar_title_text="x3"),

                        width=350,
                        height=350,
    ))


    fig.show()
```

```python
# a)
num = (7*s+1)*(3*s+1)**2
display(num.expand())

den = (5*s+1)*(4*s+1)*(s+1)
display(den.expand())

num = [63, 51, 13, 1]
den = [20, 29, 10, 1]

Visualisation(num, den, 10)
```

```python
# b)
num = (s+1)
display(num.expand())

den = (5*s**2+s+1)*(.5*s+1)
display(den.expand())

num = [1, 1]
den = [2.5, 5.5, 1.5, 1]

Visualisation(num, den,50)
```

```python
# c)
den = (5*s**2+s+1)*(25*s+1)
display(den.expand())

num = [1, 1]
den = [125, 30, 26, 1]

Visualisation(num, den, 70)
```

```python
# d)
den = (4*s+1)*(6*s+1)
display(den.expand())

num = [-3]
den = [24, 10, 1]

Visualisation(num, den, 30)
```

```python
# e)
num = (-2*s+1)*(-3*s+1)
display(num.expand())

den = (.5*s+1)*(.4*s+1)*(.1*s)
display(den.expand())

num = [6, -5, 1]
den = [.02, .09, 0.1]

Visualisation(num, den,5)
```

```python
# g)
num = (s+1)
display(num.expand())

den = (4*s+1)*(s-1)
display(den.expand())

num = [1, 1]
den = [4, -3, -1]

Visualisation(num, den)
```
